```python
import pandas as pd

income = "../data/income.csv"
df = pd.read_csv(income)
df = df[~((df["EmpStatus"] == '?' ) | (df["Origin"] == '?' ) | (df["Occupation"] == '?' ))]
```

```python
empStatus = pd.get_dummies(df["EmpStatus"])
edu = pd.get_dummies(df["Edu"])
maritalStatus = pd.get_dummies(df["MaritalStatus"])
occupation = pd.get_dummies(df["Occupation"])
familyStatus = pd.get_dummies(df["FamilyStatus"])
race = pd.get_dummies(df["Race"])
sex = pd.get_dummies(df["Sex"])
origin = pd.get_dummies(df["Origin"])

data = pd.concat(
    [
        empStatus,
        edu,
        maritalStatus,
        occupation,
        familyStatus,
        race,
        sex,
        origin
    ],
    axis=1
)
```

```python
data = pd.concat(
    [
        data,
        df["Weight"],
        df["Age"],
        df["EduYears"],
        df["CapitalGain"],
        df["CapitalLoss"],
        df["Hours"],
        df["IncomeId"]
    ],
    axis=1
)
```
